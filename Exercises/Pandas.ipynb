### loc vs iloc

```python
# .loc uses values in [] as labels
df.loc[2:6]
```

```python
# .loc uses values in [] as positions 
df.iloc[2:6]
```

```python
df.loc[:, 'species']
```

```python
df.iloc[:, 4]
```

```python
df.iloc[2]
```

```python
df.iloc[2:4, 1]
```

```python
df.iloc[::-10]
```
