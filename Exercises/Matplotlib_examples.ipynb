```python
import matplotlib.pyplot as plt
import numpy as np
```

## Intro to pyplot

```python
plt.plot([1, 2, 3, 4])
plt.ylabel('some numbers')
plt.show()
```

```python
plt.plot([1, 2, 3, 4], [1, 4, 9, 16])
```

### Formatting the style of your plot

```python
plt.plot([1, 2, 3, 4], [1, 4, 9, 16], 'ro')
plt.axis([0, 6, 0, 20])
plt.show()
```

### Plotting 2 different plots on the one axis

```python
x = np.arange(0.0, 10.0, 0.01)
y = 3.0 * x + 1.0
```

```python
noise = np.random.normal(0.0, 1.0, len(x))
```

```python
plt.plot(x, y+noise, 'r.')
plt.plot(x, y, 'b-')
plt.show()
```

### Titles, legends and labels

```python
plt.plot(x, y+noise, 'r.', label="Actual")
plt.plot(x, y, 'b-', label="Model")


plt.title("Simple plot")
plt.xlabel("Weight")
plt.ylabel("Mass")
plt.legend()

plt.show()
```

### Histograms

```python
plt.hist(noise, bins=15)
```

### Plotting side-by-side

```python
plt.subplot(1,2,1)
x = np.random.normal(0.0, 1.0, 10000)
plt.hist(x)
plt.axis([-3.5, 3.5, 0, 3000])

plt.subplot(1,2,2)
x = np.random.uniform(-3.0, 3.0, 10000)
plt.hist(x)
plt.axis([-3.5, 3.5, 0, 3000])

plt.show()
```

### Scatter plot

```python
n = 100
x = np.random.uniform(0.0, 10.0, n)
y = np.random.uniform(0.0, 100.0, n)
z = np.random.normal(100, 40.0, n)
c = np.random.randint(0, 20, n)

plt.scatter(x, y, c=c, s=z)
plt.show()
```

### Trigonometric functions

```python
x = np.arange(-2.0 * np.pi, 2.0 * np.pi, 0.1)

plt.plot(x, np.sin(x), 'g.')
plt.plot(x, np.cos(x), 'b.')

plt.show()
```

### Power of x

```python
x = np.arange(0.0, 16, 0.1)

plt.plot(x, x**2, 'g-', label="x^2")
plt.plot(x, x**3, 'b-', label="x^3")
plt.plot(x, x**4, 'r-', label="x^4")
plt.plot(x, 2**x, 'k-', label="2^x")

plt.legend()
plt.show()
```
